### What is data?

Data is anything that has meaning and the computer usually stores it in ones and zeros.

Data types allows the computer to know what it can and can't do with the data stored in the variable

### Data Types In Python
| Data Type | Classes | Description |
|-----------|----------|--------------|
| Numeric   | int,float,complex| holds numeric values |
| String    | str     | holds text data  |
| Sequence  | list, tuple,range | holds a collection of items |
| Mapping   | dict    | stores data in key-value pairs |
| Set       | set,frozenset     | unordered or  mutable collections of unique elements |
| Boolean   | bool    | represents True or False value |


#### 1 Numeric Data Types
- Integers, floating-point numbers and complex numbers falls under numbers.

    
  1) ```int``` : Represents an integer value
  2)  ```float```: Holds decimal point numbers and is accurate upto 15decimal places
  3) ```complex```: Holds complex numbers

- We use **```type()```** function to know which data type a variable or function  belongs to.

Lets see an example:

`````python
num1  = 5
print(num1, "is which data type? ", type(num1))

num2 = 2.1
print(num2, "is  which data type? ", type(num2))

num3 = 1+2j
print(num3, "is of which data type? ", type(num3))
`````

#### 2 List Data Type

List is an ordered collection of similar or different data types separated by commas and enclosed within  square brackets **[]**. 

For example:

````python
    languages = ["Python", "Java", "C++"]
````

##### Accessing List Items
- Indexes are used to access list items. -The index of the first item is `0`,
- For examples,

```python
languages = ["Python", "Java", "C++"]

#Accessing the first item
print(languages[0]) # Output: Python

#Accessing the last item
print(languages[-1]) # Output: C++

#Or
print(languages[2])  # Output: C++
```

### 3 Tuple Data Type
This is an ordered sequence of items same as a list 

The difference being  tuples are immutable. Once created , the elements cannot be changed or added to.

`````python
product = ('Xbox', 499.99)
`````

##### Accessing Tuple Items
Similar to lists, we use indexs to access tuple items






`````python
# create a tuple 
product = ('Microsoft', 'Xbox', 499.99)

# access element at index 0
print(product[0])   # Microsoft

# access element at index 1
print(product[1])   # Xbox
`````

### 4 String Data Type

- String is a sequence  of characters represented with single or double quotes
- It can contain letters, digits, special characters and numbers
- For example;

`````python
name = 'Microsoft'
print(name) #Output: Microsoft

message = "I need 4  apples."
print(message)  # Output: I need 4 apples.
`````

### 5 Set Data Type
- Set is an unordered collection  of items stored in a hash table. 
- It does not allow duplicate members, so the order will be  different each time you run it.
- It is defined by values separated by commas inside braces  `{}`.
- For example,

`````python
# create a set named student_id
student_id = {112, 114, 116, 118, 115}

# display student_id elements
print(student_id)
#Output: {112, 114, 115, 116, 118}


# display type of student_id
print(type(student_id))
# Output: <class 'set'>
`````

### 6 Dictionary Data Type
- This is an ordered collection of items stored in a hash table. 
- Each item consists of a key/value pair. 
- A key being a unique identifier associated with each value.



`````python
# create a dictionary named capital_city
capital_city = {'Nepal': 'Kathmandu', 'Italy': 'Rome', 'England': 'London'}

print(capital_city)

#Output
{'Nepal': 'Kathmandu', 'Italy': 'Rome', 'England': 'London'}
`````

- In this example, we created a dictioary named `capital_city` that contains the capital city of each country in Europe
1) **Keys** are `Nepal`, `Italy` `England` 
2) **Values** are `Kathmandu`, `Rome`, `London`

##### Accessing Items in a dictionary
- We use keys to retrieve its respective value and not the other way around.
- For example,

`````python
# create a dictionary named capital_city
capital_city = {'Nepal': 'Kathmandu', 'Italy': 'Rome', 'England': 'London'}

print(capital_city['Nepal'])  # prints Kathmandu

print(capital_city['Kathmandu'])  # throws error message 
`````
